# Bag of words technique

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"text":["likhith like sports",
                         "I watch sports",
                         "likhith plays sports",
                          "i love sports"],"output":[1,1,0,0]})

df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
#vocabulary
print(cv.vocabulary_)
```

```python
bow.toarray()
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
# new
cv.transform(['I watch sports']).toarray()
```

```python
X = bow.toarray()
y = df['output']
```

#  N-grams

```python
df = pd.DataFrame({"text":["likhith like sports",
                         "I watch sports",
                         "likhith plays sports",
                          "i love sports"],"output":[1,1,0,0]})

df
```

```python
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
#Ti gram
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(3,3))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

# TF-IDF (Term frequency- Inverse document frequency)

```python
df = pd.DataFrame({"text":["likhith like sports",
                         "I watch sports",
                         "likhith plays sports",
                          "i love sports"],"output":[1,1,0,0]})

df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid= TfidfVectorizer()
```

```python
arr = tfid.fit_transform(df['text']).toarray()
```

```python
arr
```
